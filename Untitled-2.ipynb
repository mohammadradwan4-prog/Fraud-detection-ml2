```python
Q1=df['transaction_amount'].quantile(0.25)
Q3=df['transaction_amount'].quantile(0.75)
IQR=Q3-Q1
df['transaction_amount']<Q1-1.5 * IQR
df['transaction_amount']>Q3+1.5 * IQR
#اكتشاف ال outliers
```

```python
x=df[['transaction_amount','device_trust_score']]
y=df['is_fraud']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
forest=RandomForestClassifier(n_estimators=50,random_state=42,class_weight='balanced')
```

```python
forest.fit(x_train,y_train)
```

```python
forest.predict([[79.11,0.323]])
```

```python
y_perd=forest.predict(x_test)
y_perd
```

```python
accuracy=accuracy_score(y_test,y_perd)
precision=precision_score(y_test,y_perd)
recall=recall_score(y_test,y_perd)
f1=f1_score(y_test,y_perd)
rco=roc_curve(y_test,y_perd)
print('accuracy:',accuracy)
print('precision:',precision)
print('recall:',recall)
print('f1:',f1)
print('rco:',rco)
```
